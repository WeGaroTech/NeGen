### **Importing Pandas**

```python
import pandas as pd
```

### importing Bible data `web.csv`

```python
df=pd.read_csv('web.csv',skiprows=5)
df["Book Name"].unique()
```

### Selecting only rows with Book Name "John"

```python
john=df[df["Book Name"]=='John']
john
```

### sorting verses properly

```python
john=john.sort_values(["Chapter","Verse"])
john
```

### Taking only the text column

```python
verses=john["Text"].tolist()
print(verses[:10])
```

### writing into en_clean.txt: one verse per line

```python
with open("en_clean.txt","w",encoding="utf-8") as f:
    for verse in verses:
        f.write(verse.strip()+"\n")
print("Cleaned Up: Total Verse:",len(verses))
```
